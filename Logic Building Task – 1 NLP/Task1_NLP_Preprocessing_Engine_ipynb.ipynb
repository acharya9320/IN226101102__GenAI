# Task 1: Conceptual Understanding



## Q1. What is the difference between "Love" and "love" in NLP?

**Answer:**

In Natural Language Processing (NLP), text is processed as tokens, and words with different letter cases are treated as different tokens. Therefore, "Love" and "love" are considered separate words.

For example, in the sentences "I Love coding" and "I love coding", both words will be stored differently in the vocabulary.

This can lead to:
- Increase in vocabulary size  
- Redundant or duplicate features  
- Reduced model efficiency  

To solve this problem, a preprocessing step called **case normalization** is applied, where all text is converted into lowercase.

Example:  
"Love" → "love"

This ensures consistency and improves model performance.



## Q2. What happens if stopwords are not removed?

**Answer:**

Stopwords are commonly used words such as "is", "the", "and", which usually do not add much meaning to the text.

If stopwords are not removed:
- The dataset becomes noisy  
- Important words lose importance  
- Model training becomes less efficient  
- Computational cost increases  

Example:  
"I am very happy"  

Without removing stopwords → ['I', 'am', 'very', 'happy']  
After removing stopwords → ['happy']  

Thus, removing stopwords helps in focusing on meaningful information.



## Q3. Provide two real-world scenarios where removing stopwords can be harmful

**Answer:**

### 1. Sentiment Analysis

In sentiment analysis, some stopwords are very important for understanding the correct meaning.

Example:  
"I am not satisfied"  

If the word "not" is removed, the sentence becomes:  
"I am satisfied"  

This changes the sentiment completely and gives incorrect results.



### 2. Question Answering Systems

In question-answering systems, stopwords help in understanding the structure and intent of the question.

Example:  
"Where is the nearest hospital?"  

Removing words like "where" or "is" can make the sentence unclear and affect the system's understanding.



## Q4. Explain the difference between stemming and lemmatization

**Answer:**

Stemming and lemmatization are techniques used to reduce words to their base form.

- **Stemming:**  
  It removes suffixes using simple rules and may produce non-dictionary words.  
  Example: "studies" → "studi"

- **Lemmatization:**  
  It uses linguistic rules and dictionaries to convert words into their correct base form.  
  Example: "running" → "run", "better" → "good"



### Difference Table

| Feature | Stemming | Lemmatization |
|--------|--------|-------------|
| Approach | Rule-based | Dictionary-based |
| Output | May be invalid | Always meaningful |
| Speed | Faster | Slower |
| Accuracy | Less | More |



## Conclusion

Stemming is faster but less accurate, while lemmatization provides more meaningful and accurate results. Therefore, lemmatization is preferred in real-world NLP applications where accuracy is important.

# Task 2: Advanced Preprocessing Function

In this task, we build a preprocessing function that cleans real-world noisy text data.

The function performs the following steps:

- Convert text to lowercase
- Remove URLs and email patterns
- Remove numbers
- Handle repeated characters (e.g., "soooo" → "so")
- Remove special characters and emojis
- Remove extra spaces
- Tokenize the text
- Remove very short words (length ≤ 2), except "no" and "not"

This helps in converting raw text into clean and meaningful tokens for NLP tasks.

```python
import re
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def preprocess_text(text):

    # Handle empty input
    if not text or not isinstance(text, str):
        return [], ""

    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # 3. Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1+', r'\1', text)

    # 6. Remove special characters & emojis
    text = re.sub(r'[^\w\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. Tokenization
    tokens = word_tokenize(text)

    # 9. Remove short tokens (≤2 except 'no', 'not')
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    # Final cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
# Testing
text = "Soooo happy!!! Visit https://test.com now 123 😊"

tokens, clean = preprocess_text(text)

print("Tokens:", tokens)
print("Cleaned Sentence:", clean)
```

# Task 3: Stress Testing

In this step, we test our preprocessing function on different types of real-world text data.

The test cases include:
- Emojis
- Numbers
- Slang / repeated characters
- URLs
- Mixed case text

This helps to check whether our preprocessing function can handle noisy and unstructured data properly.

```python
# Stress testing the preprocessing function on multiple real-world inputs
# Test sentences
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Apply preprocessing
results = []

for s in sentences:
    tokens, clean = preprocess_text(s)

    print("\nOriginal Text:", s)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)

    results.append(tokens)
```

# Task 4: Token Analytics

In this step, we analyze the tokens generated after preprocessing.

For each sentence, we calculate:
- Total number of tokens
- Number of unique tokens
- Average token length

This helps us understand how much noise was removed and how meaningful the processed text is.

```python
import numpy as np

# Token analysis
for i, tokens in enumerate(results):

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    print(f"\nSentence {i+1}")
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
```

## Analysis

**1. Which sentence had the most noise?**

The sentence "Win $$$ now!!! Limited offer!!!" had the most noise because it contained multiple special characters, repeated punctuation, and promotional text.

---

**2. Which sentence retained the most meaningful tokens after cleaning?**

The sentence "I absolutely looooved this product 😍😍" retained the most meaningful tokens after preprocessing, as it clearly represents a strong sentiment even after cleaning.

# Task 5: Frequency Analysis

In this step, we combine all tokens from the processed sentences and analyze their frequency.

We use the Counter class from the collections module to:
- Find the most frequent words
- Find the least frequent words

This helps in identifying important and commonly used terms in the dataset.

```python
from collections import Counter

# Combine all tokens into one list
all_tokens = [word for sublist in results for word in sublist]

# Count frequency
counter = Counter(all_tokens)

# Top 10 frequent words
print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

# Least frequent words
print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

## Analysis

The most frequent words represent common patterns in the dataset and may indicate important keywords.

The least frequent words represent rare or unique terms that appear only once or very few times.

Frequency analysis helps in understanding word importance and can be useful for feature selection in machine learning models.

# Task 6: Full Pipeline

In this step, we build a complete NLP preprocessing pipeline that processes multiple sentences at once.

The pipeline:
- Takes a list of raw text inputs
- Applies preprocessing to each sentence
- Returns combined tokens and cleaned sentences

This makes the system reusable and scalable for real-world applications.

```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
output = full_pipeline(sentences)

print("All Tokens:\n", output["tokens"])
print("\nCleaned Sentences:\n", output["clean_sentences"])
```
